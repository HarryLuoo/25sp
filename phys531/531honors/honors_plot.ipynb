1D dirac equation dispersion plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
v = 1.0  # Velocity (set to 1 for simplicity)
m_gapped = 1.0  # Mass for the gapped case
m_gapless = 0.0 # Mass for the gapless case (Dirac point)

# Momentum range
p = np.linspace(-4, 4, 400) # Adjust range and points as needed

# Calculate energies
E_plus_gapped = np.sqrt(m_gapped**2 + (v * p)**2)
E_minus_gapped = -E_plus_gapped

E_plus_gapless = np.sqrt(m_gapless**2 + (v * p)**2) # = v * |p|
E_minus_gapless = -E_plus_gapless

# --- Plotting ---
plt.figure(figsize=(7, 5))

# Plot gapped case (m=1)
plt.plot(p, E_plus_gapped, 'b-', label=f'$m={m_gapped}$')
plt.plot(p, E_minus_gapped, 'b-')

# Plot gapless case (m=0)
plt.plot(p, E_plus_gapless, 'r--', label=f'$m={m_gapless}$')
plt.plot(p, E_minus_gapless, 'r--')

# Add horizontal lines for the gap
plt.axhline(m_gapped, color='gray', linestyle=':', linewidth=0.8)
plt.axhline(-m_gapped, color='gray', linestyle=':', linewidth=0.8)
# Add text for gap energy values
plt.text(p[-1]*0.8, m_gapped + 0.1, f'$E=m$', color='gray', fontsize=9)
plt.text(p[-1]*0.8, -m_gapped - 0.2, f'$E=-m$', color='gray', fontsize=9)

# Add labels and title
plt.xlabel('$p$ (Momentum)')
plt.ylabel('$E(p)$ (Energy)')
plt.title('Dispersion Relation for 1D Dirac Hamiltonian ($v=1$)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.axhline(0, color='black', linewidth=0.5) # Zero energy line
plt.axvline(0, color='black', linewidth=0.5) # Zero momentum line

# Set y-axis limits for better visualization (optional)
plt.ylim(-E_plus_gapped.max()*1.1, E_plus_gapped.max()*1.1)

# Save the plot (choose one format)
plt.savefig('1d_dirac_dispersion.png', dpi=300)
# plt.savefig('1d_dirac_dispersion.pdf')

# Show the plot
plt.show()
```

2D dirac eq dispersion


```python
import numpy as np
import matplotlib.pyplot as plt
# Import Axes3D specifically for 3D plotting (even if not directly used, needed for projection='3d')
from mpl_toolkits.mplot3d import Axes3D # noqa

# Parameters
v = 1.0  # Velocity (set to 1 for simplicity)
m_gapped = 1.0  # Mass for the gapped case
m_gapless = 0.0 # Mass for the gapless case (Dirac cone)

# Momentum grid
px_range = np.linspace(-3, 3, 60) # Increased points for smoother surface
py_range = np.linspace(-3, 3, 60)
Px, Py = np.meshgrid(px_range, py_range)
P = np.sqrt(Px**2 + Py**2)

# Calculate energies
E_plus_gapped = np.sqrt(m_gapped**2 + (v * P)**2)
E_minus_gapped = -E_plus_gapped

E_plus_gapless = np.sqrt(m_gapless**2 + (v * P)**2) # = v * P
E_minus_gapless = -E_plus_gapless

# --- Plotting Gapped Case (m=1) ---
fig_gapped = plt.figure(figsize=(7, 6))
ax_gapped = fig_gapped.add_subplot(111, projection='3d', azim=-60, elev=20) # Adjusted view angle

# Plot surfaces
surf_gapped_plus = ax_gapped.plot_surface(Px, Py, E_plus_gapped, cmap='viridis', alpha=0.9, rstride=1, cstride=1, linewidth=0.1, antialiased=True)
surf_gapped_minus = ax_gapped.plot_surface(Px, Py, E_minus_gapped, cmap='cividis', alpha=0.9, rstride=1, cstride=1, linewidth=0.1, antialiased=True)

# Labels and Title
ax_gapped.set_xlabel('$p_x$')
ax_gapped.set_ylabel('$p_y$')
ax_gapped.set_zlabel('$E(\mathbf{p})$')
ax_gapped.set_title(f'2D Dirac Dispersion (Gapped, $m={m_gapped}, v={v}$)')
ax_gapped.dist = 11 # Zoom out slightly
ax_gapped.zaxis.set_major_locator(plt.MultipleLocator(1)) # Adjust Z ticks

# Save
plt.savefig('2d_dirac_dispersion_gapped.png', dpi=300, bbox_inches='tight')

# --- Plotting Gapless Case (m=0) ---
fig_gapless = plt.figure(figsize=(7, 6))
ax_gapless = fig_gapless.add_subplot(111, projection='3d', azim=-60, elev=20) # Adjusted view angle

# Plot surfaces (Dirac cone)
surf_gapless_plus = ax_gapless.plot_surface(Px, Py, E_plus_gapless, cmap='plasma', alpha=0.9, rstride=1, cstride=1, linewidth=0.1, antialiased=True)
surf_gapless_minus = ax_gapless.plot_surface(Px, Py, E_minus_gapless, cmap='magma', alpha=0.9, rstride=1, cstride=1, linewidth=0.1, antialiased=True)

# Labels and Title
ax_gapless.set_xlabel('$p_x$')
ax_gapless.set_ylabel('$p_y$')
ax_gapless.set_zlabel('$E(\mathbf{p})$')
ax_gapless.set_title(f'2D Dirac Dispersion (Gapless, $m={m_gapless}, v={v}$)')
ax_gapless.dist = 11 # Zoom out slightly
ax_gapless.zaxis.set_major_locator(plt.MultipleLocator(1)) # Adjust Z ticks


# Save
plt.savefig('2d_dirac_dispersion_gapless.png', dpi=300, bbox_inches='tight')

# Show plots (optional, will show one after the other)
plt.show()
```
